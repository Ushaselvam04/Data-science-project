```python
#import packages
import seaborn
import matplotlib.pyplot as plt
#loading dataset using seaborn
df=seaborn.load_dataset('tips')
#df.head()
print(df.to_string())
seaborn.pairplot(df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df= sns.load_dataset('tips')
#selected_vars=['total_bill','tip']
#sns.pairplot(df,vars=selected_vars)
#seaborn.pairplot(df,hue='size',c='color#')
#seaborn.pairplot(df,hue="size",palette='husl')
#sns.pairplot(df,diag_kind='kde')

#adjusting regression
sns.pairplot(df,kind='reg')

plt.show()
```

```python
#controlling the markers

sns.pairplot(df,hue='sex',markers=["o","s"])
plt.show()
```

```python
# Advanced costomization with seaborn pairplot

g=sns.pairplot(df, hue='day')
g.fig.suptitle("Pairplot of tips Dataset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
